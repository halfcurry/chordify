```python
def read_pkls():
	data = []
	labels = []
	files = os.listdir(os.path.join(os.getcwd(),'data'))
	files = sorted(files)
	print(files)
	for file in files:
		X = joblib.load(os.path.join(os.getcwd(),'data',file))
		data.append(X)
	files = os.listdir(os.path.join(os.getcwd(),'labels'))
	files = sorted(files)
	print(files)
	for file in files:
		Y = joblib.load(os.path.join(os.getcwd(),'labels',file))
		labels.append(Y)
	return data, labels
```
